## <font color="#07a8ed">**Conjuntos**

En matemáticas, un conjunto es una colección de elementos bien definidos (es decir, se puede determinar con claridad si un elemento pertenece o no al conjunto) y distintos entre sí (no hay elementos repetidos).

En Python, existe una estructura de datos llamada `set` que permite trabajar con conjuntos de forma sencilla y eficiente.

* Los elementos de un conjunto se escriben entre llaves `{}` separados por comas.

* Los conjuntos no admiten duplicados.

* Un conjunto vacío se crea con la función `set()`.


```python
A = {1, 2, 3}
A
```

```python
B = set() # conjunto vacio
```

```python
C = {2, 4, 6}
C
```

```python
type(A)
```

Los objetos de tipo `set` en Python son estructuras mutables, lo que significa que pueden modificarse después de ser creados. Es posible agregar elementos utilizando el método add() y eliminarlos mediante el método discard().

Por ejemplo, al conjunto $A$ definido previamente se le puede añadir el elemento 6, mientras que del conjunto $C$ se puede eliminar el elemento 4:

```python
A.add(6)
C.discard(4)
```

## <font color="#07a8ed">**Pertenencia**

La pertenencia hace referencia a la relación que indica si un elemento está contenido en un conjunto.

En matemáticas, esta relación se expresa mediante los símbolos:

* $a$ ∈ $A$ → el elemento $a$ pertenece al conjunto $A$.

* $a$ ∉ $A$ → el elemento $a$ **no** pertenece al conjunto $A$.


En Python, la pertenencia a un conjunto se evalúa mediante los operadores `in` y `not in`.

El operador in retorna el valor lógico `True` si el elemento pertenece al conjunto y `False` en caso contrario.

El operador `not in` funciona de manera inversa: devuelve `True` si el elemento no pertenece al conjunto y `False` si pertenece.

```python
1 in A
```

```python
4 in A
```

```python
4 not in A
```

```python
1 not in A
```

## <font color="#07a8ed">**Subconjuntos**

Un conjunto $A$ es subconjunto de $B$ si todos los elementos de $A$ pertenecen también a $B$. Se expresa mediante:  

$$A \subseteq B$$  

En Python, esta relación se puede verificar mediante la función `issubset()`. La instrucción `A.issubset(B)` devuelve `True` si $A$ es subconjunto de $B$, y `False` en caso contrario.

```python
A = {1, 2}
B = {1, 2, 3, 4, 5}

print("¿A es subconjunto de B?", A.issubset(B))
```

```python
C = {4, 6}
D = {1, 2, 3, 4, 5}

print("¿C es subconjunto de D?", C.issubset(D))
```

## <font color="#07a8ed">**Operaciones con conjuntos**

Las operaciones básicas entre conjuntos son:

### <font color="#07a8ed">**1️⃣ Unión**

La unión de conjuntos es una operación que combina los elementos de dos conjuntos distintos en uno solo. El resultado es un conjunto que contiene todos los elementos que pertenecen a, por lo menos, uno de los conjuntos.

En matemáticas se expresa como:

$$A \cup B = \{ x \mid x \in A \; \text{o} \; x \in B \}$$

En Python, la unión se puede realizar con el operador `|` o mediante la función `union()`.

Union.jpg

```python
A = {1, 2, 3, 4}
B = {2, 4, 6, 8}
A|B
```

```python
A = {'Hola', 'Mundo'}
B = {'Hola', 'a', 'todos'}
A.union(B)
```

### <font color="#07a8ed">**2️⃣ Intersección**


a intersección de conjuntos es una operación que devuelve un nuevo conjunto formado únicamente por los elementos que son comunes a ambos conjuntos.

En matemáticas se representa como:

$$A \cap B = \{ x \mid x \in A \; \text{y} \; x \in B \}$$

En Python, la intersección se realiza con el operador `&` o mediante la función `intersection()`.

intersección.jpg

```python
A&B
```

```python
A.intersection(B)
```

### <font color=#07a8ed>**3️⃣ Diferencia**


La diferencia de conjuntos es una operación que devuelve un nuevo conjunto formado por los elementos que pertenecen a
𝐴
A pero no a
𝐵
B.

En matemáticas se expresa como:

$$A - B = \{ x \mid x \in A \; \text{y} \; x \notin B \}$$

En Python, la diferencia se puede calcular con el operador `-` o mediante la función `difference()`.

diferencia.jpg

```python
A-B
```

```python
A.difference(B)
```

### <font color="#07a8ed">**4️⃣ Diferencia simétrica**

La diferencia simétrica entre dos conjuntos es la operación que devuelve un nuevo conjunto formado por los elementos que pertenecen a uno de los conjuntos, pero no a ambos simultáneamente.

En matemáticas se representa como:

$$A \triangle B = \{ x \mid (x \in A \; \text{o} \; x \in B) \; \text{y} \; x \notin (A \cap B) \}
$$

En Python, la diferencia simétrica se realiza con el operador `^` o mediante la función `symmetric_difference()`.


diferencia_simétrica.jpg

```python
A^B
```

```python
A.symmetric_difference(B)
```
