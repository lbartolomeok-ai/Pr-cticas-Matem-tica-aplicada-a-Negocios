## <font color="#07a8ed">**Proposición**

Una proposición es una afirmación que puede ser verdadera (`True`) o falsa (`False`), pero nunca ambas a la vez.

Ejemplos de proposiciones:

* “El número 10 es mayor que 5” (Verdadera).

* “La Tierra es plana” (Falsa).

```python
P = True   # "10 > 5"
Q = False  # "La Tierra es plana"

print("P:", P)
print("Q:", Q)
```

## <font color="#07a8ed">**Conectores lógicos**

Los conectores lógicos permiten combinar proposiciones:

* Negación (¬P / not P): invierte el valor.

* Conjunción (P ∧ Q / P and Q): es verdadera solo si ambas lo son.

* Disyunción (P ∨ Q / P or Q): es verdadera si al menos una lo es.

* Condicional (P → Q): solo es falsa si P es verdadera y Q es falsa.

* Bicondicional (P ↔ Q): es verdadera si P y Q tienen el mismo valor.

### <font color="#07a8ed">**1️⃣ Negación (¬P)**

La negación devuelve falso cuando es aplicada sobre un verdadero y viceversa. Es decir, invierte el valor de verdad de la proposición.

| P     | ¬P    |
|-------|-------|
| True  | False |
| False | True  |

En Python: `not` P

```python
not False
```

```python
not True
```

### <font color="#07a8ed">**2️⃣ Conjunción (P ∧ Q)**


La conjunción devuelve verdadero si ambas proposiciones son verdaderas. O lo que es lo mismo, si alguna de sus proposiciones es falsa devuelve falso.

| P     | Q     | P ∧ Q |
|-------|-------|-------|
| True  | True  | True  |
| True  | False | False |
| False | True  | False |
| False | False | False |

En Python: P `and` Q

```python
True and True
```

```python
True and False
```

```python
False and False
```

```python
False and True
```

### <font color=#07a8ed>**3️⃣ Disyunción (P ∨ Q)**


La disyunción devuelve verdadero si al menos una proposición es verdadero. O lo que es lo mismo, devuelve falso cuando ambas preposiciones son falsas.

| P     | Q     | P ∨ Q |
|-------|-------|-------|
| True  | True  | True  |
| True  | False | True  |
| False | True  | True  |
| False | False | False |

En Python: P `or` Q

```python
True or False
```

```python
True or True
```

```python
False or True
```

```python
False or False
```

### <font color="#07a8ed">**4️⃣ Condicional (P → Q)**

La proposición condicional ("si P entonces Q") es falsa solo cuando P es verdadera y Q es falsa, en cualquier otro caso es verdadera.

Esta proposición no afirma que P ocurra, afirma que si P ocurre, entonces Q ocurrirá.

Si P es falsa, el condicional siempre se considera verdadero, porque no se violó la afirmación: “si P, entonces Q”. Esto se conoce como “verdadero por vacuidad”.


| P     | Q     | P → Q |
|-------|-------|-------|
| True  | True  | True  |
| True  | False | False |
| False | True  | True  |
| False | False | True |

En Python: `(not P) or Q`

```python
(not True) or True
```

```python
(not True) or False
```

```python
(not False) or True
```

```python
(not False) or False
```

### <font color="#07a8ed">**5️⃣ Bicondicional (P ↔ Q)**

El bicondicional (“P si y solo si Q”) es verdadero cuando P y Q tienen el mismo valor de verdad y falso cuando difieren.

También se puede ver como la conjunción de dos condicionales:

$$P \leftrightarrow Q \equiv (P \rightarrow Q) \wedge (Q \rightarrow P)$$

| P     | Q     | P ↔ Q |
|-------|-------|-------|
| True  | True  | True  |
| True  | False | False |
| False | True  | False |
| False | False | True |

En Python: P == Q

```python
True == True
```

```python
True == False
```

```python
False == False
```

```python
False == True
```

### <font color="#07a8ed">**1️⃣Estructuras condicionales `if` - `elif` - `else`**

Permiten ejecutar diferentes bloques de código según una condición (verdadera o falsa).

## <font color="#07a8ed">**Cuantificadores**

En la lógica, además de las proposiciones simples y los conectores lógicos, aparecen los cuantificadores, que permiten generalizar afirmaciones sobre conjuntos de elementos.

Existen principalmente dos tipos:



### <font color="#07a8ed">**1️⃣Cuantificador universal (∀)**

Se lee: “para todo” o “para cada”.
Expresa que una propiedad se cumple para todos los elementos de un conjunto.

Ejemplo en matemática:

$$\forall n \in \mathbb{N}, \; n + 0 = n$$

Esto significa: para todo número natural $n$, si sumamos 0 el resultado es el mismo número.

En Python: `all()`

```python
# Verificar que para todo n en una lista se cumple n + 0 = n
numeros = [0, 1, 2, 3, 10, 100]
resultado = all(n + 0 == n for n in numeros)
print(resultado)
```

### <font color="#07a8ed">**2️⃣Cuantificador existencial (∃)**

Se lee: “existe al menos uno”.
Expresa que hay al menos un elemento en el conjunto para el cual se cumple la propiedad.

Ejemplo en matemática:

$$\exists n \in \mathbb{N}, \; n^2 = 4$$

Esto significa: existe un número natural $n$ tal que su cuadrado es 4.

En Python: `any()`

```python
# Verificar si existe al menos un número en la lista cuyo cuadrado sea 4
numeros = [0, 1, 2, 3, 10, 100]
resultado = any(n**2 == 4 for n in numeros)
print(resultado)  # True (porque n = 2)
```

## <font color="#07a8ed">**Negación de cuantificadores**

La **negación de cuantificadores** permite transformar afirmaciones universales en existenciales y viceversa. Esto es fundamental en lógica matemática y en programación.

### <font color="#07a8ed">**1️⃣Negación del cuantificador universal (∀)**

Regla:

$$\neg (\forall x \; P(x)) \equiv \exists x \; \neg P(x)$$

Interpretación:

* "No es cierto que todos los elementos cumplen $P(x)$".

* Equivale a: "Existe al menos un elemento que no cumple $P(x)$".

Ejemplo matemático:

* Afirmación: “Todos los números de la lista son positivos”.

$$\forall n \in \text{lista}, \; n > 0$$

* Negación: “Existe al menos un número que no es positivo”.

$$\exists n \in \text{lista}, \; n \leq 0$$




```python
numeros = [3, 5, -2, 7]

# Cuantificador universal: ¿todos son positivos?
todos_positivos = all(n > 0 for n in numeros)
print("Todos positivos?", todos_positivos)

# Negación equivalente: ¿existe al menos uno que no sea positivo?
existe_no_positivo = any(n <= 0 for n in numeros)
print("Alguno no positivo?", existe_no_positivo)
```

### <font color="#07a8ed">**2️⃣Negación del cuantificador existencial (∃)**

Regla:

$$\neg (\exists x \; P(x)) \equiv \forall x \; \neg P(x)$$

Interpretación:

* "No existe ningún elemento que cumpla $P(x)$".

* Equivale a: "Para todo elemento, $P(x)$ no se cumple".

Ejemplo matemático:

* Afirmación: “Existe al menos un número en la lista que es divisible por 7”.

$$\exists n \in \text{lista}, \; 7 \mid n$$

* Negación: “Todos los números de la lista NO son divisibles por 7”.

$$\forall n \in \text{lista}, \; 7 \nmid n$$

```python
numeros = [3, 5, 10, 14]

# Cuantificador existencial: ¿existe un número divisible por 7?
existe_div_7 = any(n % 7 == 0 for n in numeros)
print("Existe divisible por 7?", existe_div_7)

# Negación equivalente: ¿todos NO son divisibles por 7?
todos_no_div_7 = all(n % 7 != 0 for n in numeros)
print("Todos no divisibles por 7?", todos_no_div_7)
```
